```python
import warnings
warnings.filterwarnings("ignore")
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
IQR = q3-q1
lower = q1 - 1.5*(IQR)
higher = q3 + 1.5*(IQR)
cleaned_data = df[~((df < lower) | (df > higher)).any(axis=1)]
cleaned_data.head()
```

```python
df1 = cleaned_data
df1.shape # After removing Outliers the shape is this...
```

```python
y = df["Price"]
y # dependent
```

```python
df.drop("Price",axis=1,inplace=True)
df.drop("Locality",axis=1,inplace=True)
x = df
x.head() # independent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=90)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = lr.predict(x_test)
print(r2_score(y_test,y_pred))
```

**Linear Regression : 62.36%**

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(x_train,y_train)
```

```python
y_pred1 = ridge.predict(x_test)
print(r2_score(y_test,y_pred1))
```

**Ridge : 62.36%**

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
```

```python
y_pred = rfr.predict(x_test)
print(r2_score(y_test,y_pred))
```

**RandomForestRegressor : 89.78%**

```python
from sklearn.tree import DecisionTreeRegressor
dtc = DecisionTreeRegressor()
dtc.fit(x_train,y_train)
```

```python
y_pred = dtc.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
parameter = {
    'criterion' : ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    'splitter' : ["best", "random"],
    'max_features':["auto", "sqrt", "log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
gsc = GridSearchCV(dtc,param_grid=parameter,scoring = 'accuracy',cv=5)
gsc.fit(x_train,y_train)
```

```python
y_pred = gsc.predict(x_test)
print(r2_score(y_test,y_pred))
```

**Decision tree regression : 81.75%**
